# Palyginkime vyrus ir moteris naudodami permutacijos metodą

Permutacijos metodas, tai pasiimame dvi grupes ir padarome bendrą imtį. Tada ją permutuojame atsitiktinai, ir iš bendros grupės pasiimame simuliuotų vyrų grupę ir simuliuotų moterų grupę. Užfiksuojame skirtumą ir jį išsaugojame histogramoje. Pakartojame kelis tūkstančius kartų ir gaunama nulinės hipotezės ($H_0$) histogramą. Žiūrėdami kur mūsų gautas skirtumas randasi $H_0$ histogramoje, galime apskaičiuoti kiek tikėtinas mūsų gautas skirtumas tarp dviejų grupių, galiojant $H_0$.

```python
vyrai   = dns_df[dns_df['LYTIS_1'] == 2.0]
moterys = dns_df[dns_df['LYTIS_1'] == 1.0]
```

DNS *sumos* palyginimas permutacijomis
-------------------------------------


```python
x = vyrai['DNS_suma'].to_list()
y = moterys['DNS_suma'].to_list()

orginalus_skirtumas = np.mean(y) - np.mean(x)
orginalus_skirtumas
```

Permutacijų metodu išsiaiškiname kiek skiriasi DNS moterų ir vyrų vidurkiai.
Remiamės H0 hipoteze, jog vidurkių skirtumų nėra ir iš jos sukuriame histogramą, tada žiūrimę kurioje vietoje histogramos patenka mūsų skirtumas. Jei patenka į ekstremumą, tada išties galima drąsiai tvirtinti, jog imtys yra iš skirtingų populiacijų, bet jeigu nėra labai didelio skirtumo, tai sunku teigti, jog tikrai vienaip ar kitaip.

```python
def formatuoti_teksta(tekstas, zodziu_eiluteje=10):
    words = tekstas.split(' ')
    lines = []
    current_line = []
    for word in words:
        current_line.append(word)
        if len(current_line) >= zodziu_eiluteje:
            lines.append(' '.join(current_line))
            current_line = []
    # Įtraukiam likusius žodžius, jei yra
    if current_line:
        lines.append(' '.join(current_line))
    formatted_text = '\n'.join(lines)
    return formatted_text

def palygint(stulpelis='DNS_suma', STEPS = 10000, ax = None, title = '', tekstas = '', fontsize=10):
    x = vyrai[stulpelis].to_list()
    y = moterys[stulpelis].to_list()
    orginalus_skirtumas = np.mean(y)-np.mean(x)


    statistic = (lambda x, y: np.mean(y) - np.mean(x))
    output = scipy.stats.permutation_test((x,y),statistic = statistic,\
                                      n_resamples = STEPS, permutation_type='independent')

    # Plotinimas
    if ax == None:
        fig,ax = plt.subplots(1,1, figsize=(10,5))

    ax.hist(output.null_distribution, bins = STEPS//100, histtype='step',alpha = 0.5)
    ax.set_title(title)
    ax.set_xlabel('atsakymų vertės')
    ax.set_ylabel('dažnis')

    # Piešiame rodyklę, rodančią vidurkių skirtumą ant histogramos
    ax.annotate(f"gautas = {orginalus_skirtumas:.2f} ir p = {output.pvalue:1.3f}",
            xy=(orginalus_skirtumas, 0), xytext=(orginalus_skirtumas, STEPS/30),
            arrowprops=dict(facecolor='Yellow', shrink=0.05),
            ha='center')  # horizontalalignment


    formatted_text = formatuoti_teksta(tekstas)  # Pavyzdžiui, eilutėje norime matyti 10 žodžių

    fig.text(0.9, 0.5, formatted_text, ha='left', va='center', fontsize=fontsize)
    plt.savefig(title+'.png')

    return output
```

```python
#thanks chatgpt!
from scipy.stats import spearmanr

# Tarkime, dns_df yra tavo DataFrame

dalykai = ['laimingumo testas', 'DNS_suma', 'pasitenkinimas vaikyste', 'tėvų religijos išlaikymas']

# Funkcija, kuri grąžina Spearmano koreliacijos koeficientų ir p reikšmių matricas
def spearman_corr_pval(df, features):
    corr = pd.DataFrame(index=features, columns=features)
    pval = pd.DataFrame(index=features, columns=features)
    for i in features:
        for j in features:
            corr_result = spearmanr(df[i], df[j])
            corr.loc[i, j] = corr_result.correlation
            pval.loc[i, j] = corr_result.pvalue
    return corr.astype(float), pval.astype(float)

# Gauname koreliacijos ir p reikšmių matricas
koreliacija, p_reiksmes = spearman_corr_pval(dns_df, dalykai)

# Vizualizuojame koreliacijos koeficientus
plt.figure(figsize=(10, 8))
sns.heatmap(koreliacija, annot=True, fmt=".2f", linewidths=.5, cbar=True)
plt.title('Koreliacijos matrica')
plt.show()

# Vizualizuojame p reikšmes
plt.figure(figsize=(10, 8))
sns.heatmap(p_reiksmes, annot=True, fmt=".2g", linewidths=.5, cbar=True, cmap="coolwarm")
plt.title('P reikšmių matrica')
plt.show()
```
